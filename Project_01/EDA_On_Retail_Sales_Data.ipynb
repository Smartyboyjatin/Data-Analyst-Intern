```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('retail_sales_dataset.csv')
```

```python
dataset.head(3)
```

```python
dataset.drop(columns=['Date'], inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.dropna(inplace=True)
```

# Highest Salling Month

```python
dataset['Month'].value_counts()
```

# Monthly Order Count 

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Month' , data=dataset)
plt.title('Monthly Order Count')
plt.savefig('01_Monthly Sale count.jpg')
plt.show()
```

# Highest Selling Product Category

```python
dataset['Product Category'].value_counts()
```

```python
x= dataset['Product Category'].value_counts().index
y= dataset['Product Category'].value_counts().values
```

```python
plt.figure(figsize=(4,3))
plt.pie(y, labels=x, autopct="%0.2f%%")
plt.title('Sale based on prodect category')
plt.savefig("02_Count of product category")
# plt.legend(loc=4)
plt.show()
```

# Order Count Based on Gender

```python
dataset['Gender'].value_counts()
```

```python
plt.figure(figsize=(4,3))
sns.countplot(x='Gender', data=dataset)
plt.title('Order Count Based on Gender')
plt.savefig('03_Order Count Based on Gender.jpg')
plt.show()
```

# Product category classification Based on Gender.

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Product Category', data=dataset, hue='Gender')
plt.title('Category Sale based on Gender')
plt.savefig('04_Category sale based on gender.jpg')
plt.show()
```

# Product Category Based on Salling Amount

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Product Category', y='Total Amount', data=dataset, estimator=sum)
plt.title('Product Category Based on Salling Amount')
plt.savefig('05_Product Category Based on Salling Amount.jpg')
plt.show()
```

# Monthly Sale Based on Total Amount

```python
plt.figure(figsize=(12,5))
sns.barplot(x='Month', y='Total Amount', data=dataset, estimator=sum)
plt.title('Monthly Sale Based on Amount')
plt.savefig('06_Monthly Sale Based on Amount.jpg')
plt.show()
```
